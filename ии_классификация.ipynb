# **Убираем пустоты:**
Удаляем строки, где неизвестно лекарство, тк мы не можем угадать какое используется:

```python
train.dropna(subset=['Drug'], inplace=True)
train.info()
train.isnull().sum()
```

Заполняем оставшиеся пропуски средними значениями:

```python
train.Cholesterol = train.Cholesterol.fillna(train.Cholesterol.median())
train.Tryglicerides = train.Tryglicerides.fillna(train.Tryglicerides.median())
train.Copper = train.Copper.fillna(train.Copper.median())
train.Platelets = train.Platelets.fillna(train.Platelets.median())
train.isnull().sum()
```

```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
train[['Status', 'Drug', 'Sex','Ascites','Hepatomegaly','Spiders','Edema']] = enc.fit_transform(train[['Status', 'Drug', 'Sex','Ascites','Hepatomegaly','Spiders','Edema']])
train.head()
```

# **Визуализация**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 5)

#Распределение классов
plt.subplot(1, 2, 1)
train['Status'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#99ff99','#ff9999'], startangle=90)
plt.title('Процентное соотношение исходов (Status)')

#Количество пациентов по полу и статусу (Count Plot)
plt.subplot(1, 2, 2)
sns.countplot(x='Status', hue='Sex', data=train, palette='pastel')
plt.title('Количество исходов в разрезе пола')
plt.show()

#Признак: Bilirubin
#Плотность распределения (KDE Plot)
plt.subplot(1, 2, 1)
sns.histplot(train['Bilirubin'], kde=True, color='orange')
plt.title('Распределение Билирубина (Гистограмма + KDE)')

#Сравнение Билирубина по Стадиям (Boxplot)
plt.subplot(1, 2, 2)
sns.boxplot(x='Stage', y='Bilirubin', data=train, palette='Set2')
plt.yscale('log') # Логарифмическая шкала, так как разброс значений велик
plt.title('Билирубин по стадиям')
plt.show()

#Признак: Stage
#Количество пациентов на разных стадиях (Bar Plot)
plt.subplot(1, 2, 1)
sns.countplot(x='Stage', data=train, palette='viridis')
plt.title('Количество пациентов на каждой стадии')

#Стадия в зависимости от наличия асцита (Stacked Bar)
plt.subplot(1, 2, 2)
stage_ascites = pd.crosstab(train['Stage'], train['Ascites'], normalize='index')
stage_ascites.plot(kind='bar', stacked=True, ax=plt.gca(), color=['#aec6cf', '#ffb347'])
plt.title('Наличие асцита в зависимости от стадии (%)')
plt.show()

#Признак: Age
train['Age_years'] = train['Age'] / 365

#Зависимость возраста от статуса (KDE по группам)
plt.subplot(1, 2, 2)
sns.kdeplot(data=train, x='Age_years', hue='Status', fill=True, common_norm=False, palette='bright')
plt.title('Плотность распределения возраста для разных исходов')
plt.show()
```

```python
#Тепловая карта корреляций
plt.figure(figsize=(12, 10))
numeric_cols = ['Age', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper', 'Alk_Phos', 'SGOT', 'Platelets', 'Prothrombin']
sns.heatmap(train[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Корреляция биохимических показателей')
plt.show()

#Сравнение Билирубина по Стадиям и Препаратам
plt.figure(figsize=(10, 6))
sns.boxenplot(x='Stage', y='Bilirubin', hue='Drug', data=train)
plt.yscale('log')
plt.title('Зависимость уровня Билирубина от Стадии и типа Лекарства')
plt.show()

#Pairplot
g = sns.pairplot(train, vars=['Bilirubin', 'Albumin', 'Prothrombin'], hue='Status', palette='bright')

g.fig.suptitle('Взаимосвязь ключевых факторов риска по исходам', y=1.03, fontsize=16)

plt.show()
```
